```python
import random

rooms = {"A": "dirty", "B": "dirty", "C": "dirty"}

clean_count = 0

current_room = "A"

while True:
    print(f"\nAgent is in room {current_room}")
    print(f"Room states: {rooms}")


    if rooms[current_room] == "dirty":
        print(f"Cleaning room {current_room}...")
        rooms[current_room] = "clean"
        clean_count += 1
    else:
      
        if all(state == "clean" for state in rooms.values()):
            current_room = random.choice(list(rooms.keys()))
            print("All rooms clean! Moving randomly to", current_room)
        else:
          
            next_rooms = list(rooms.keys())
            current_index = next_rooms.index(current_room)
            current_room = next_rooms[(current_index + 1) % len(next_rooms)]

    
    if all(state == "clean" for state in rooms.values()):
        print("\nAll rooms are clean!")
        print("Total times cleaned:", clean_count)
        break
```

```python
def smart_light_agent(room_occupied, time_of_day, natural_light):
    """
    room_occupied: True/False
    time_of_day: "day" or "night"
    natural_light: "bright" or "dim"
    """
    
    if room_occupied and (time_of_day == "night" or natural_light == "dim"):
        action = "Turn ON light"
    elif not room_occupied:
        action = "Turn OFF light"
    else:
        action = "Keep light OFF"

    return action


print(smart_light_agent(True, "night", "dim"))   # Turn ON
print(smart_light_agent(False, "day", "bright")) # Turn OFF
print(smart_light_agent(True, "day", "bright"))  # Keep OFF
```

```python
import random


current_temp = 22
past_temps = [] 

for step in range(10):
    print(f"\nTime step {step+1}")
    print(f"Current temperature: {current_temp}°C")

  
    past_temps.append(current_temp)
    if len(past_temps) > 3:
        past_temps.pop(0)

    
    if len(past_temps) == 3:
        trend = past_temps[-1] - past_temps[0]
    else:
        trend = 0

   
    if current_temp < 20:
        action = "Turn ON heater"
    elif current_temp > 24:
        action = "Turn ON cooler"
    elif trend > 1:
        action = "Cooling trend detected – prepare to cool"
    elif trend < -1:
        action = "Heating trend detected – prepare to heat"
    else:
        action = "Temperature stable – no action"

    print("Agent action:", action)

    
    current_temp += random.choice([-1, 0, 1])
```
